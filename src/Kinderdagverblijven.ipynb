### Dan laden we wat data

De data komt van <http://www.landelijkregisterkinderopvang.nl/opendata/export_opendata_lrkp.csv>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Hier komt het volgende stukje

```python
kdv = pd.read_csv('datafiles/export_opendata_lrkp.csv', sep=";", parse_dates=[6,7], low_memory=False)
```

```python
kdv_indatum = kdv.groupby("inschrijfdatum").agg({"actuele_naam_oko": pd.Series.nunique})
```

```python
inschrijvingen = kdv_indatum.plot()
```

```python
kdv_uitdatum = kdv.groupby("uitschrijfdatum").agg({"actuele_naam_oko": pd.Series.nunique})
```

```python
uitschrijvingen = kdv_uitdatum[-365:].plot()
```
